For DB2, we will be importing two libraries- `ibm_db` and `ibm_dbi`.   
`ibm_db` is a library with low-level functions that will directly connect to our db2 database.   
To make things easier, we will be using `ibm-dbi`, which communicates with ibm-db and gives us an easy interface to interact with our data and import our data as a pandas dataframe.  
If the library is not installed, we need to install the library first before importing.

````python
import pandas as pd
import ibm_db
import ibm_db_dbi
````

Connect to a Db2 database
--
You might sense what to do already. Yes, replace the necessarry part such as DATABASE, HOSTNAME, PORT, UID and PWD.   

````python
# replace only <> credentials
dsn = "DRIVER={{IBM DB2 ODBC DRIVER}};" + \
      "DATABASE=SAMPLE;" + \
      "HOSTNAME=lamps1.fyre.ibm.com;" + \
      "PORT=50008;" + \
      "PROTOCOL=TCPIP;" + \
      "UID=db2v1155;" + \
      "PWD=password;"
hdbc  = ibm_db.connect(dsn, "", "")
hdbi = ibm_db_dbi.Connection(hdbc)
````
